```python
#Detect the outliers in Salary column using IQR method
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
print(Q1)
print(Q3)
print(IQR)
```

```python
LB = Q1 - 1.5 * IQR
UB = Q3 + 1.5 * IQR
print(LB)
print(UB)
```

```python
outliers = df[(df['Salary'] < LB) | (df['Salary'] > UB )]
print("Outliers in the salary column")
print(outliers)
```

```python
#Detect outliers of age column using Z-score method


import numpy as np
#Load the dataset
import pandas as pd
df = pd.read_csv('/content/sample_data/employee_data (1).csv')
df.head()
mean1 = df["Salary"].mean()
std1 = df["Salary"].std()
```

```python
def zscore(data, mean1, std1):
  res = (data-mean1)/std1
  return res
```

```python
df["Age1"] = df["Age"]
```

```python
for i in range(len(df["Age"])):
  df["Age"][i] = zscore(df["Age"][i], mean1, std1)
```

```python
#Load the Datasets
#Load the dataset
import pandas as pd
df = pd.read_csv('/content/sample_data/employee_data (1).csv')
df.head()
df.info()
```

```python
def check_duplicated_rows(data) :
  seen = set() #Unique rows
  duplicates = [] #List to store duplicate rows

  for row in data:
    row_tuple = tuple(row) #Convert from list to tuple
    if row_tuple in seen:
      duplicates.append(row)
    else:
      seen.add(row_tuple)

  return duplicates
```

```python
data = [
    [1,"Shravya",54],
    [2,"Cheery",48],
    [3,"Tom",26],
    [1,"Shravya",54],
]
```

```python
duplicates = check_duplicated_rows(data)
duplicates
```
